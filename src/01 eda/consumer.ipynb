```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
consumer_df = pd.read_csv("../../data/intermediate_data/consumer_city.csv")
```

### Age-wise Preferred Flavor
Apple is the most preferred flavor across most of the age groups. Though, Orange is the most selling flavor.

```python
consumer_df.groupby(["Age_Group", "Preferred_Flavor"])["Customer_ID"].count().sort_values(ascending=False)
```

```python
# Visualize the relationship between Age_Group and Preferred_Flavor
plt.figure(figsize=(14, 8))
sns.countplot(data=consumer_df, x='Age_Group', hue='Preferred_Flavor')
plt.title('Age Group vs Preferred Flavor')
plt.show()
```

### Preferred Channel
Interestingly, E-commerce is the most preferred channel, though the sales are highest for General trade.

```python
consumer_df.groupby(["Preferred_Channel"])["Customer_ID"].count().sort_values(ascending=False)
```

### Distribution of age group

```python
# Visualize the distribution of Age_Group
plt.figure(figsize=(10, 6))
sns.countplot(data=consumer_df, x='Age_Group', order=consumer_df['Age_Group'].value_counts().index)
plt.title('Distribution of Age Groups')
plt.show()
```

### Distribution of Income group
1. The graph indicates that the majority of individuals have relatively low income levels, while a smaller subset of individuals have significantly higher income levels.
2. The right-skewed nature of the distribution suggests economic disparity, where a large portion of the population earns less, and only a few individuals fall into the higher income brackets.

```python
# Visualize the distribution of Income_Level
plt.figure(figsize=(10, 6))
sns.histplot(data=consumer_df, x='Income_Level', bins=30, kde=True)
plt.title('Distribution of Income Levels')
plt.show()
```

### Relationship between Preferred_Channel and Purchase_Frequency
1. General Trade tends to have slightly higher purchase frequencies and a broader range of customer behavior.
2. E-commerce, Modern Trade, HoReCa, and Q-commerce exhibit similar patterns with consistent purchase frequencies, although there are occasional outliers who shop more frequently.
3. The presence of outliers in all channels highlights a subset of customers with significantly higher purchase activity.

```python
# Visualize the relationship between Preferred_Channel and Purchase_Frequency
plt.figure(figsize=(14, 8))
sns.boxplot(data=consumer_df, x='Preferred_Channel', y='Purchase_Frequency')
plt.title('Preferred Channel vs Purchase Frequency')
plt.show()
```

### Price Sensitivity vs Purchase Frequency
Price Sensitivity does not appear to have a strong impact on the median purchase frequency or the overall range of purchase frequency.

```python
# Visualize the relationship between Price Sensitivity and Purchase Frequency
plt.figure(figsize=(14, 8))
sns.boxplot(data=consumer_df, x='Price_Sensitivity', y='Purchase_Frequency')
plt.title('Price Sensitivity vs Purchase Frequency')
plt.show()
```
